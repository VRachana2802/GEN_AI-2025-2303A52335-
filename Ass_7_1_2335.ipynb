```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adadelta

np.random.seed(42)
num_samples = 1000
X = np.random.rand(num_samples, 8)
X[:, 0] = X[:, 0] * 10
X[:, 1] = X[:, 1] * 150 + 50
X[:, 2] = X[:, 2] * 100
X[:, 3] = X[:, 3] * 50
X[:, 4] = X[:, 4] * 500
X[:, 5] = X[:, 5] * 50
X[:, 6] = X[:, 6] * 2.5
X[:, 7] = X[:, 7] * 80 + 20

y = np.random.randint(0, 2, size=num_samples)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

model = Sequential([
    Dense(8, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(16, activation='relu'),
    Dense(20, activation='relu'),
    Dense(10, activation='relu'),
    Dense(1, activation='sigmoid')
])

model.compile(
    optimizer=Adadelta(),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

history = model.fit(X_train, y_train, epochs=150, batch_size=64, validation_data=(X_test, y_test))
model.save('diabetes_diagnosis_model.h5')
y_pred_prob = model.predict(X_test)
y_pred = (y_pred_prob > 0.5).astype(int)

accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

print(f"Test Accuracy: {accuracy}")
print("Confusion Matrix:")
print(conf_matrix)
print("Classification Report (Precision, Recall, F1-score):")
print(class_report)
loaded_model = load_model('diabetes_diagnosis_model.h5')
sample = X_test[0].reshape(1, -1)
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adadelta

np.random.seed(42)
num_samples = 1000
X = np.random.rand(num_samples, 8)
X[:, 0] = X[:, 0] * 10
X[:, 1] = X[:, 1] * 150 + 50
X[:, 2] = X[:, 2] * 100
X[:, 3] = X[:, 3] * 50
X[:, 4] = X[:, 4] * 500
X[:, 5] = X[:, 5] * 50
X[:, 6] = X[:, 6] * 2.5
X[:, 7] = X[:, 7] * 80 + 20

y = np.random.randint(0, 2, size=num_samples)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

model = Sequential([
    Dense(8, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(16, activation='relu'),
    Dense(20, activation='relu'),
    Dense(10, activation='relu'),
    Dense(1, activation='sigmoid')
])

model.compile(
    optimizer=Adadelta(),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

history = model.fit(X_train, y_train, epochs=150, batch_size=64, validation_data=(X_test, y_test))
model.save('diabetes_diagnosis_model.h5')
y_pred_prob = model.predict(X_test)
y_pred = (y_pred_prob > 0.5).astype(int)

accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

print(f"Test Accuracy: {accuracy}")
print("Confusion Matrix:")
print(conf_matrix)
print("Classification Report (Precision, Recall, F1-score):")
print(class_report)
loaded_model = load_model('diabetes_diagnosis_model.h5')
sample = X_test[0].reshape(1, -1)
predicted_disease_prob = loaded_model.predict(sample)
predicted_class = (predicted_disease_prob > 0.5).astype(int)
print(f"Predicted Diabetes (0: No, 1: Yes): {predicted_class[0][0]}")
```
